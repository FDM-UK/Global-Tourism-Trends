# UNWTO Tourist Information

We are using Tourism Information from the UN, downloaded as an xls spreadsheet, to establish countries annual number of visitor. 

The information is limited as it only goes up to 2022. It also does not include regions or cities. It does, however, differentiate between overnight visitors and day visitors and breaks down day visitors into those who arrived via cruise ship. This helps us define quiter locations for a digital nomad to work and visit cultural spots with less disruption.

#### Replacing column names and ensuring that the appropriate columns are string type


```python
df_raw.rename(columns={'Basic data and indicators': 'Country'}, inplace=True)
df_raw.rename(columns={'Unnamed: 4': 'Arrivals'}, inplace=True)
df_raw[df_raw.columns[1]] = df_raw[df_raw.columns[1]].astype(str) + ' ' + \
                                                df_raw[df_raw.columns[2]].astype(str) + ' ' + \
                                                df_raw[df_raw.columns[3]].astype(str) + ' ' + \
                                                df_raw[df_raw.columns[4]].astype(str)

columns_to_drop2 = ['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7']       
df_raw.drop(columns=columns_to_drop2, errors='ignore', inplace=True)                                        


df_raw.head(10)
```

```python
df_cleaned = df_raw
```

#### Removing Countries with no visitor data


```python
countries_to_remove = []
current_country = None
has_visitor_data = False
country_start_index = None  # Marks the start of a block of rows for a country

for index, row in df_cleaned.iterrows():
    if pd.notna(row['Country']) and row['Country'].strip():
        # Found new country

        # If the last country had no visitor data, mark its rows for removal
        if current_country is not None and not has_visitor_data:
            countries_to_remove.extend(range(country_start_index, index))

        # Update tracking for the new country
        current_country = row['Country']
        country_start_index = index
        has_visitor_data = False

    elif current_country:
        # Still inside a country block; now check if there's any visitor data in this row
        for year in range(2018, 2023):
            try:
                if pd.notna(pd.to_numeric(row[year])) and pd.to_numeric(row[year]) != 0:
                    has_visitor_data = True
                    break  # No need to keep checking if we already found data
            except (ValueError, TypeError):
                # If it can’t be turned into a number, ignore and keep going
                continue

# After the loop, check if the last country also had no data
if current_country and not has_visitor_data:
    countries_to_remove.extend(range(country_start_index, len(df_cleaned)))

# Drop those rows and tidy up the index
df_filtered = df_cleaned.drop(index=countries_to_remove).reset_index(drop=True)

# Show the first 20 rows to check our result
print(df_filtered.head(20))
```

```python
df_filtered['Country'] = df_filtered['Country'].str.lower()
df_filtered['Arrivals'] = df_filtered['Arrivals'].str.lower()
df_filtered['Units'] = df_filtered['Units'].str.lower()
df_filtered
```
